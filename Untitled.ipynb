```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import random
```

```python
thetaList = np.random.uniform(-np.pi/4, np.pi/4,size=100)
rList = np.random.uniform(-10,10,size=100)
```

```python
plt.polar(thetaList,rList,marker='o',ls='')
```

```python
xList = rList * np.cos(thetaList)
```

```python
yList = rList * np.sin(thetaList)
```

```python
plt.scatter(xList,yList)
```
