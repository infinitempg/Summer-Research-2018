# Spectral Evolution of SN2010jl

```python
import aplpy
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import os
import astropy.table as Table
from astropy.io import ascii
from scipy.interpolate import interp1d
from astropy.io import fits

plt.rcParams.update({'font.size':12})
```

```python
path = './SN_Images/SN2010jl/'
```

```python
def plotSpec(path,file,center):
    file = fits.open(path+file)
    lwl = file[0].header['CRVAL1']
    delW = file[0].header['CD1_1']
    width = file[0].header['NAXIS1']
    plt.figure(figsize=(10,7.5))
    plt.plot(np.linspace(lwl,len(file[0].data) * delW + lwl,width)-center,file[0].data)
    plt.axvline(0,ls='--',c='gray')
    plt.xlim(-150,150)
    plt.show()
    return((np.linspace(lwl,len(file[0].data) * delW + lwl,width)-center,file[0].data))
```

```python
sn2012 = plotSpec(path,'SN2010jl_01_B_535.ms.fits',6621.43)
```
