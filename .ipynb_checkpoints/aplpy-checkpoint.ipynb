```python
import aplpy
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import os
import astropy.table as Table
from astropy.io import ascii
from scipy.interpolate import interp1d
```

```python
def losVel(func, lam, apCen):
    return (func(apCen)-lam)/func(apCen) * 3e5
```

```python
offset = 5177.5-800
```

```python
SN07Kdata = ascii.read('SN_Images/SN07K/SN2007K_01_B.csv')
```

```python
SN07Kdata['Adjusted Aperture'] = SN07Kdata['Aperture Center'] + offset
```

```python
SN07Kdata
```

```python
apSpace = np.linspace(800,940,141)
f = interp1d(SN07Kdata['Aperture Center'],SN07Kdata['Peak (Gaussian)'],kind='cubic')
fm = interp1d(SN07Kdata['Aperture Center'],SN07Kdata['Peak (Maximum)'],kind='cubic')
```

```python
gaussArr = np.vstack((f(apSpace),apSpace+offset))
maxArr = np.vstack((fm(apSpace),apSpace+offset))
```

```python
def addPeak(apCent,peak,figname,right=False,pm=5,col='r',fontcol='w'):
    if right:
        sign=1
        align='left'
    elif not right:
        sign=-1
        align='right'

    figname.add_label(peak+3*sign,apCent+offset,r'$\lambda = %s \AA$'%peak,horizontalalignment=align,color=fontcol)
#     figname.show_arrows(peak,apCent+offset,0,pm,head_width=2,head_length=0,width=0.5,length_includes_head=True,color=col)
#     figname.show_arrows(peak,apCent+offset,0,-pm,head_width=2,head_length=0,width=0.5,length_includes_head=True,color=col)
    return
```

```python
fig = plt.figure(figsize=(15,9))

SN07Kg = aplpy.FITSFigure('SN_Images/SN07K/SN2007K_01_B.fits',figure=fig,subplot=[0.05,0.05,0.425,0.9])
SN07Kg.recenter(6695,5251,width=50,height=160)
SN07Kg.show_grayscale(stretch='power',pmax=99)
SN07Kg.set_title('SN2007K - Gaussian - %.2f km/s Offset'%losVel(f,6705.27,872.3))
SN07Kg.axis_labels.set_xtext('Wavelength (Angstroms)')
SN07Kg.axis_labels.set_ytext('Aperture Center (pix)')
SN07Kg.show_markers(gaussArr[0,:],gaussArr[1,:],marker='.',edgecolor='orange')
for line in SN07Kdata:
    addPeak(line['Aperture Center'],line['Peak (Gaussian)'],SN07Kg)
addPeak(872,6705.27,SN07Kg,pm=10,col='k',fontcol='k')
SN07Kg.show_markers(SN07Kdata['Peak (Gaussian)'],SN07Kdata['Adjusted Aperture'],marker='o',s=50,edgecolor=None,facecolor='r')
SN07Kg.show_markers(6705.27,872.3+offset,marker='*',s=150,edgecolor=None,facecolor='r')

SN07Km = aplpy.FITSFigure('SN_Images/SN07K/SN2007K_01_B.fits',figure=fig,subplot=[0.55,0.05,0.425,0.9])
SN07Km.recenter(6695,5251,width=50,height=160)
SN07Km.show_grayscale(stretch='power',pmax=99)
SN07Km.set_title('SN2007K - Maximum - %.2f km/s Offset'%losVel(fm,6701.237,872.3))
SN07Km.axis_labels.set_xtext('Wavelength (Angstroms)')
SN07Km.axis_labels.set_ytext('Aperture Center (pix)')
SN07Km.show_markers(maxArr[0,:],maxArr[1,:],marker='.',edgecolor='orange')
for line in SN07Kdata:
    addPeak(line['Aperture Center'],line['Peak (Maximum)'],SN07Km)
addPeak(872,6701.237,SN07Km,right=False,pm=10,col='k',fontcol='k')
SN07Km.show_markers(SN07Kdata['Peak (Maximum)'],SN07Kdata['Adjusted Aperture'],marker='o',s=50,edgecolor=None,facecolor='r')
SN07Km.show_markers(6701.237,872.3+offset,marker='*',s=150,edgecolor=None,facecolor='r')

fig.tight_layout()
fig.savefig('SN_Images/SN07K/SN07K.png')
```
