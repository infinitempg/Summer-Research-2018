```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.io.ascii as ascii
import astropy.table as Table
from scipy.interpolate import interp1d
from scipy import interpolate
```

```python
SN07Kdata = ascii.read('SN_Images/SN2007K_01_B.csv')
SN07Kdata
```

```python
apSpace = np.linspace(800,940,141)
f = interp1d(SN07Kdata['Aperture Center'],SN07Kdata['Peak (Gaussian)'],kind='cubic')
fm = interp1d(SN07Kdata['Aperture Center'],SN07Kdata['Peak (Maximum)'],kind='cubic')

# tck = interpolate.splrep(SN07Kdata['Aperture Center'],SN07Kdata['Peak (Gaussian)'])
# f_S = interpolate.splev(apSpace,tck)
# tck2 = interpolate.splrep(SN07Kdata['Aperture Center'],SN07Kdata['Peak (Maximum)'])
# fm_S = interpolate.splev(apSpace,tck2)
```

```python
def losVel(func, lam, apCen):
    return (func(apCen)-lam)/func(apCen) * 3e5
```

```python
(f(872.3)-6705.27)/f(872.3) * 3e5,(fm(872.3)-6701.237)/fm(872.3) * 3e5
```

```python
fig = plt.figure(figsize=(14,7))

fig.add_subplot(1,2,1)
plt.title('SN_2007K - Gaussian - %.2f km/s'%losVel(f,6705.27,872.3))
plt.xlabel('Aperture Center (px)')
plt.ylabel(r'H$\alpha$ Peak Wavelength ($\AA$)')

plt.scatter(SN07Kdata['Aperture Center'],SN07Kdata['Peak (Gaussian)'])
plt.plot(apSpace,f(apSpace),'--',label='interpolation')
plt.scatter(872.3,6705.27,label='SN2007K')
plt.legend()

fig.add_subplot(1,2,2)
plt.title('SN_2007K - Maximum - %.2f km/s'%losVel(fm,6701.237,872.3))
plt.xlabel('Aperture Center (px)')
plt.ylabel(r'H$\alpha$ Peak Wavelength ($\AA$)')

plt.scatter(SN07Kdata['Aperture Center'],SN07Kdata['Peak (Maximum)'])
plt.plot(apSpace,fm(apSpace),'--',label='interpolation')
plt.scatter(872.3,6701.237,label='SN2007K')
plt.legend()

plt.tight_layout()
plt.savefig('SN_Images/SN07K/SN07K_interp.png')
```
