```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.stats import norm, cauchy
from scipy.optimize import minimize
```

```python
xspace = np.linspace(0,100,101)
```

## 1 gaussian

```python
def loss(P,x,y):
    loc,scale,weight = P
    yhat = weight*cauchy.pdf(x,loc,scale)
    return ((y - yhat)**2).sum()
```

```python
P_real = [55.5,5.64,19.64]
```

```python
testNorm = cauchy.pdf(xspace,loc=P_real[0],scale=P_real[1])
testNoise = 1/20*norm.rvs(loc=0,scale=0.1,size=101)
testData = P_real[2]*(testNorm + testNoise)
```

```python
P_guess = [40,3,10]
```

```python
res = minimize(loss,P_guess,args=(xspace,testData),method='Nelder-Mead')
```

```python
P_fit = res['x']
```

```python
fitData = P_fit[2]*cauchy.pdf(xspace,loc=P_fit[0],scale=P_fit[1])
```

```python
res
```

```python
plt.plot(testData)
# plt.plot(testNorm,'--')
plt.plot(fitData,'k')
```

## 2 gaussian

```python
# loc1,scale1,weight1,loc2,scale2,weight2,weightall
p2 = [35,5,20,40,20,20]
```

```python
testNorm1 = cauchy.pdf(xspace,loc=p2[0],scale=p2[1])
testNorm2 = cauchy.pdf(xspace,loc=p2[3],scale=p2[4])
testNoise = norm.rvs(loc=0,scale=0.1,size=101)
testData2g = (p2[2]*testNorm1 + p2[5]*testNorm2 + testNoise)
```

```python
plt.plot(testData2g)
plt.plot(p2[2]*testNorm1)
plt.plot(p2[5]*testNorm2)
```

```python
def loss2(P,x,y):
    loc1,scale1,weight1,loc2,scale2,weight2 = P
    yhat = weight1*cauchy.pdf(x,loc1,scale1) + weight2*cauchy.pdf(x,loc2,scale2)
    return ((y - yhat)**2).sum()
```

```python
P_guess2 = [40,3,10,15,15,6]
```

```python
res2 = minimize(loss2,P_guess2,args=(xspace,testData2g),method='Nelder-Mead')
```

```python
P_fit2 = res2['x']
```

```python
P_fit2
```

```python
fitData2 = P_fit2[2]*cauchy.pdf(xspace,loc=P_fit2[0],scale=P_fit2[1]) + P_fit2[5]*cauchy.pdf(xspace,loc=P_fit2[3],scale=P_fit2[4])
```

```python
plt.figure(figsize=(15,7))
plt.plot(testData2g)
plt.plot(fitData2,'--')
plt.plot(P_fit2[2]*cauchy.pdf(xspace,loc=P_fit2[0],scale=P_fit2[1]),alpha=0.5)
plt.plot(P_fit2[5]*cauchy.pdf(xspace,loc=P_fit2[3],scale=P_fit2[4]),alpha=0.5)
```
