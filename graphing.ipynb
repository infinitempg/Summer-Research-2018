```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.facecolor'] = 'w'
```

```python
from astropy.table import Table
```

```python
HAlpha = 6562.8
```

```python
def makeGraphs(dataLines,name):
    dataRows = [line for line in dataLines]
    t = Table(names=('B','R','Peak'),rows=dataRows)
    
    fig = plt.figure(figsize=(12,5))
    fig.add_subplot(1,2,1)
    plt.title('%s - B Aperture vs Peak'%name)
    plt.scatter(t['B'],t['Peak'])
    plt.xlabel('B Aperture')
    plt.ylabel(r'H$\alpha$ Peak Wavelength (Angstroms)')
    
    fig.add_subplot(1,2,2)
    plt.title('%s - R Aperture vs Peak'%name)
    plt.scatter(t['R'],t['Peak'])
    plt.xlabel('R Aperture')
    plt.ylabel(r'H$\alpha$ Peak Wavelength (Angstroms)')
    
    plt.tight_layout()
    plt.savefig('images/%s.png'%name)
    plt.show()
    
    print([getZ(line[-1]) for line in dataRows])
    return t
```

```python
def getZ(lamObs):
    return (lamObs-HAlpha)/HAlpha
```
